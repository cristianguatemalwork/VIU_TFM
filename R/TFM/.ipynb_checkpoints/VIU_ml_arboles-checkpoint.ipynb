```python
#DBSCAN
minPts=2
features_norm1 = features_norm

A = kneighbors_graph(features_norm1, minPts, include_self=False)
Ar = A.toarray()

seq = []
for i,s in enumerate(features_norm1):
    for j in range(len(features_norm1)):
        if Ar[i][j] != 0:
            seq.append(matdist[i][j])

seq.sort()
# establecer intervalo ejes
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(np.arange(0, 120, 10))
ax.set_yticks(np.arange(1, 0.5, 0.1))
plt.plot(seq)

x_dado = 45  # Reemplaza 5.0 con el valor de X deseado

# Valores de X y Y para la interpolación
x_valores = np.arange(0, len(seq))  # Valores de X son los índices de la secuencia
y_valores = seq

# Realizar interpolación lineal
interp_func = interp1d(x_valores, y_valores)

# Encontrar el valor de Y correspondiente a x_dado
y_correspondiente = interp_func(x_dado)

print(f"Para X = {x_dado}, Y = {y_correspondiente}")

# Agregar una línea horizontal para el valor de y correspondiente
plt.axhline(y=y_correspondiente, color='red', linestyle='--', label=f'Y={y_correspondiente}')

plt.annotate(f'Y={y_correspondiente}', xy=(x_dado, y_correspondiente), xytext=(x_dado, y_correspondiente + 0.000001),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=10, color='red')

plt.show()
```

```python
#KKN
from sklearn.preprocessing import StandardScaler
col_na = data_knn.columns
# Hacer el split de los datos
train, test = train_test_split( data_knn, test_size = 0.2, random_state = semilla )

# Normalizar solo el conjunto de entrenamiento
scaler = StandardScaler()
val_no =['SEXO','PRES','N_PRI_P', 'N_PUB_P', 'N_IND_P', 'N_VOL_EC_P','N_VOL_EX_P', 'M_PRI_P',
         'M_PUB_P', 'M_IND_P', 'M_VOL_EC_P','M_VOL_EX_P','TARGET']
X_train = scaler.fit_transform( train.drop( val_no, axis=1))  
# Transformar el conjunto de prueba usando el mismo escalador
X_test = scaler.transform( test.drop( val_no, axis=1))  

train = pd.concat([ pd.DataFrame(X_train).reset_index().drop(['index'], axis=1), 
                    pd.DataFrame(train[val_no]).reset_index().drop(['index'], axis=1)], axis=1)
train.columns = col_na
test = pd.concat([ pd.DataFrame(X_test).reset_index().drop(['index'], axis=1), 
                   pd.DataFrame(test[val_no]).reset_index().drop(['index'], axis=1)], axis=1)
test.columns = col_na
```

```python
 train.drop( val_no, axis=1)
```

```python
inicio = time.time()
cv = KFold(n_splits = 2, shuffle = True, random_state = semilla )

for i, weights in enumerate(['uniform', 'distance']):
    total_scores = []
    for n_neighbors in range(1,10):
        fold_accuracy = []
        knn = neighbors.KNeighborsClassifier( n_neighbors, weights = weights, n_jobs = -1 )
        for train_fold, test_fold in cv.split(train):
            # División train test aleatoria
            f_train = train.loc[train_fold]
            f_test = train.loc[test_fold]
            # entrenamiento y ejecución del modelo
            knn.fit( X = f_train.drop(['TARGET'], axis=1),
                     y = f_train['TARGET'])
            y_pred = knn.predict(X = f_test.drop(['TARGET'], axis = 1))
            # evaluación del modelo
            acc = accuracy_score(f_test['TARGET'], y_pred)
            fold_accuracy.append(acc)
        total_scores.append(sum(fold_accuracy)/len(fold_accuracy))

    plt.plot(range(1,len(total_scores)+1), total_scores, marker='o', label=weights )
    print ('Max Value ' +  weights + " : " +  str(max(total_scores)) +" (" + str(np.argmax(total_scores) + 1) + ")")
    plt.ylabel('Acc')


fin = time.time()  
tm= fin-inicio
print('Tiempo de ejecución es:',tm//3600,'horas con',tm%3600//60,'minutos y', tm%60,'segundos' )
```

```python
n_neighbors = 2
weights = 'uniform'
knn = neighbors.KNeighborsClassifier(n_neighbors= n_neighbors, weights=weights)
# fit and predict
knn.fit(X = train.drop(['TARGET'], axis=1), y = train['TARGET'])
y_pred = knn.predict(X = test.drop(['TARGET'], axis=1))
acc =  accuracy_score(test['target'], y_pred)
print ('MAE', acc)
```
