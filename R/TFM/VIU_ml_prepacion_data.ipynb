```python
for col in datos.columns:
    missing = datos[col].isnull()
    num_missing = np.sum(missing)

    if num_missing > 0:
        datos['{}_ismissing'.format(col)] = missing


# # then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in datos.columns if 'ismissing' in col]
datos['num_missing'] = datos[ismissing_cols].sum(axis=1)
missing_counts = datos['num_missing'].value_counts().reset_index()
missing_counts.columns = ['num_missing', 'count']  # Renombrar las columnas

# Ahora puedes hacer el plot
missing_counts.sort_values(by='num_missing').plot.bar(x='num_missing', y='count')
plt.xlabel('Número de valores faltantes')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de valores faltantes por registro')
plt.show()
```

```python
#Se hace la transformación de la variable sexo
lb = preprocessing.LabelBinarizer()
datos['SEXO_T'] = datos['SEXO']
datos['SEXO'] = lb.fit_transform( datos['SEXO'])
# 0 para H y 1 para M
lb = LabelEncoder()
datos['PRES_T'] = datos['PRES']
datos['PRES'] = lb.fit_transform( datos['PRES'])
#0 para dis, 1 para inv y 2 para vej
```

```python
#Se borran las columnas adicionales creadas en el bloque anterior
#Para los datos de entrenamiento
datos1 = datos.filter(regex='^(?!.*missing)').copy()
datos = datos1

del datos1
gc.collect()
```

```python
#Se hace un proceso de limpieza de datos
print('Cedulas con fecha de nacimiento nan', datos[datos['FECHA_NACIMIENTO'].isna()]['CEDULA_COD'].nunique()) #1
print('Cedulas con pension calcula de nan', datos[datos['PENSION_CAL'].isna()]['CEDULA_COD'].nunique()) #208 
print('Cedulas con RANGO_INI_5MEJ de nan', datos[datos['RANGO_INI_5MEJ'].isna()]['CEDULA_COD'].nunique()) #206
print('Cedulas con RANGO_INI_5MEJ de nan', datos[datos['RANGO_FIN_5MEJ'].isna()]['CEDULA_COD'].nunique()) #206
print('Cedulas con LS2 de nan', datos[datos['LS2'].isna()]['CEDULA_COD'].nunique())  #15
print('Cedulas con SAL_PROM2 de nan', datos[datos['SAL_PROM2'].isna()]['CEDULA_COD'].nunique())  #15
```

```python
#Se eliminan registros que no se pueden recuperar
datos.dropna( subset=['FECHA_NACIMIENTO'], inplace=True)
datos.dropna( subset=['PENSION_CAL'], inplace=True)
```

```python
datos['RANGO_INI_5MEJ'] = np.where( datos['RANGO_INI_5MEJ'].isna(),
                                    datos['INI_CAL'],
                                    datos['RANGO_INI_5MEJ'] )

datos['RANGO_FIN_5MEJ'] = np.where( datos['RANGO_FIN_5MEJ'].isna(),
                                    datos['FIN_CAL'],
                                    datos['RANGO_FIN_5MEJ'] )

datos['LS2'] = np.where( datos['LS2'].isna(),
                         datos['LS1'],
                         datos['LS2'] )

datos['SAL_PROM2'] = np.where( datos['SAL_PROM2'].isna(),
                               datos['SAL_PROM1'],
                               datos['SAL_PROM2'] )
```

```python
#Se crea el target
today = datetime.today()
datos['TARGET'] = np.where( (datos['CJ'] + datos['KM'] + datos['DB']) == 3 , 1, 0)
#Se crea la edad del pensionistas a la fecha de derecho
datos['EDAD_J'] = (datos['FECHA_DERECHO'] - datos['FECHA_NACIMIENTO']).dt.days // 365
datos['TIEM_T'] = datos['FECHA_NACIMIENTO'].apply(lambda x: today.year - x.year - ((today.month, today.day) < (x.month, x.day)))
datos['TIEM_MA'] = (datos['RANGO_INI_5MEJ'] - datos['RANGO_FIN_5MEJ']).dt.days // 30
```

```python
datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX',
        'N_PRI', 'N_PUB', 'N_IND', 'N_VOL_EC', 'N_VOL_EX', 
        'N_COD_TR']]= datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX','N_PRI', 
                              'N_PUB', 'N_IND', 'N_VOL_EC', 'N_VOL_EX', 'N_COD_TR']].fillna(0)
```

```python
datos['M_PRI_P'] = datos['M_PRI'] / datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX']].sum(axis=1)
datos['M_PUB_P'] = datos['M_PUB'] / datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX']].sum(axis=1)
datos['M_IND_P'] = datos['M_IND'] / datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX']].sum(axis=1)
datos['M_VOL_EC_P'] = datos['M_VOL_EC'] / datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX']].sum(axis=1)
datos['M_VOL_EX_P'] = datos['M_VOL_EX'] / datos[['M_PRI', 'M_PUB', 'M_IND', 'M_VOL_EC', 'M_VOL_EX']].sum(axis=1)

datos['N_PRI_P'] = datos['N_PRI'] / datos['MES_AS']
datos['N_PUB_P'] = datos['N_PUB'] / datos['MES_AS']
datos['N_IND_P'] = datos['N_IND'] / datos['MES_AS']
datos['N_VOL_EC_P'] = datos['N_VOL_EC'] / datos['MES_AS']
datos['N_VOL_EX_P'] = datos['N_VOL_EX'] / datos['MES_AS']
datos['N_COD_TR_P'] = datos['N_COD_TR'] / datos['MES_AS']
```
